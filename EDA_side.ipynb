```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('2017-1.xls')

df
```

```python
numCol = ['Happiness.Score','Whisker.high','Whisker.low','Economy..GDP.per.Capita.','Family','Health..Life.Expectancy.','Freedom','Generosity','Trust..Government.Corruption.','Dystopia.Residual']
```

```python
df[numCol].hist(edgecolor='black', linewidth=1.2, bins=15)
fig = plt.gcf()
fig.set_size_inches(30, 50)
plt.show()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(df[numCol].corr(method='pearson'), annot=True)
plt.show
```
